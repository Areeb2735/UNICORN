```python
# compare the two txt file and find the common lines

def compare_txt(file1, file2):
    with open(file1) as f:
        lines1 = f.readlines()
    with open(file2) as f:
        lines2 = f.readlines()
    common_lines = []
    diff_lines = []
    for line in lines1:
        if line in lines2:
            common_lines.append(line)
        else:
            diff_lines.append(line)
    return diff_lines

compare_txt('filenames.txt', 'filenames_home.txt')
```
